1️⃣Challenge 1: Rick and Morty API Usage

```python
def allowed_to_collect(url):
    
    return requests.get(url=url)
```

```python
allowed_to_collect("https://rickandmortyapi.com/api/character")
```

```python
'''
hago llamado a la api depues combierto lo que me trajo la api y lo combierto a json, para poder sacar la siguiente pagina entro al sub info
y ya dentro de info y guardado busco el next para poder sacar la url de la suguiente pagina. Teniendo eso y sabiendo que en la ultima pagina el "next" 
tiene como valor none puedo usarlo de condicion, hasta que no se cumpla esa condicion no va dejar de taer y de aplicarlo en una lista 


'''
todo=[]
url_next= 'https://rickandmortyapi.com/api/character'
while True:
    response=allowed_to_collect(url_next) 
    

    if response.status_code != 200:
        print('no se pudo traer la informacion')
        break
    
    response=response.json()
   
    info = response['info']
    #print(info)
    url_next= info['next']
    #print(url_next)
       
    if url_next != None:
        print('check')
        results=response['results']
        todo.extend(results)
        
        
        
    else:
        print('acabe')
        break
    
    '''' 
    if "next" in info :
        print('check')
        results=response['results']
        next_next=response['info']['next']
        print(results)
        break
        results= response.json()['results']
        after=response.json()["paging"]["next"]["after"]
        total=results,after
        todo.append(total)
        print(after)
    '''    
        
```

2️⃣Challenge 2: Filter and transformation

```python
'''
aca lo comvierto en un dataframe t empiezo a manipularlo para identificar el estimado de la edad,y 
poder transformar los diccionarios que estan en el dataframe
'''
df=pd.DataFrame(todo)
df['longitud'] = df['episode'].apply(len)
df
df.loc[df['longitud'] >= 10, 'estimated age']='younger' #| df.loc[df['longitud'] <= 10, 'estimated age']='older'
df.loc[df['longitud'] <= 10, 'estimated age']='older'
df["origin"] = df["origin"].apply(lambda x: f"{x.get('name')} - {x.get('url')}")
df["location"] = df["location"].apply(lambda x: f"{x.get('name')} - {x.get('url')}")



df.to_csv(r"C:\Users\Juan Pablo Pelaez\Documents\Training\Week 5\Rick_and_Morty.csv",index=False)
```
